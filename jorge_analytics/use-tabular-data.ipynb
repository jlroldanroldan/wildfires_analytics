```python
# Import packages
import os 

import matplotlib.pyplot as plt
import pandas as pd

# Download and open the .csv file using Pandas
avg_monthly_precip = pd.read_csv("https://ndownloader.figshare.com/files/12710618")

# View the data that you just downloaded and opene
avg_monthly_precip
```

```python
# view stats of the dataframe 
avg_monthly_precip.info()
```

```python
# Plot the data
f, ax = plt.subplots()
avg_monthly_precip.plot(x="months",
                        y="precip",
                        title="Precipitation (mm) for One Year",
                        ax=ax)
ax.set(xlabel='Month',
       ylabel='Precipitation (mm)')
plt.show()


                        
```
